# Part 10: A Billion Here, a Trillion There
A couple of lessons ago we looked at how an individual perceptron works. It keeps track of its connections, the *weights* on those connections, and its own *bias*. We got to play around with the weights and the inputs from other neurons to see if the perceptron got triggered. This was all modeled on neurons in the brain.

Let's this time take a look at the connections.

## Back to the Squishy Bits
Think back to the brain we zoomed into before. A brain is not a flat, two-dimensional object, it is squishy in all three dimensions. And within it, the neurons themselves are three-dimensional:

* They can have connections from many many other neurons nearby, in any direction where there is space.
* But even though surrounded by other neurons, they could be mostly disconnected from their neighbors. Growing connections is part of how the brain grows and stores experiences.
* While connections are growing, the weights can be constantly changing, so that a neuron listens more or less strongly to those neighbors it is connected to.
* And its bias can change over time, becoming more or less easily triggered.
* And it can be growing its own connections to other neurons to deliver more messages when it gets triggered.

And then think that every other neuron in the brain is also receiving from its existing connections, adding new connections, changing weights and bias, all day and night. The whole system of neurons is in constant intercommunication and change.

### How Many Neurons?
Remember that our models have been using 4-8 perceptrons per color in a single layer, so a few hundred so far. If you trust [Wikipedia](https://en.wikipedia.org/wiki/List_of_animals_by_number_of_neurons), we're talking about:

* 5,600 for a jellyfish
* 100,000 for a lobster
* 250K for an ant
* 1M for a cockroach
* 70M for a mouse
* 240M for a guinea pig
* 50M for an octopus
* 760M for a cat
* 2.2B for a dog
* 5B for a lion
* 86B for a human
* 257B for an elephant

### How many connections?
Ever notice how a brain is twisty and curvy? That shape arranges the neurons to be as close to each other as possible, while still letting blood vessels get in to provide oxygen and energy.

So each neuron can only connect to a certain number of nearby neurons. That's still a lot of possible connections. And the more neurons you have, the number of possible connections from those three-dimensional neighbors continues to skyrocket.

* Human: 150,000,000,000,000
* Cat:    10,000,000,000,000
* Mouse:   1,000,000,000,000

Yes we're talking about trillions of connections here. But importantly, it's not just the number of connections, it's how the connections are made and used.

## So What About the Math?
Time to get back to our perceptrons. As noted previously, what we're using for our rather simple color matching problem is a vintage 1980s style network. It works just fine for what we're doing. Here's a diagram for 4 perceptrons per color label. The red, green, and blue inputs at the top act as sensors for the RGB values in our pixel. Each of them is fully connected to every perceptron in the middle layer. Each of the middle layer perceptrons has its own bias, and each keeps a weight for how much it listens to the R, G, and B inputs. Then at the bottom is an output perceptron for each of the color labels we are predicting. Each of these also has a bias,  each is connected to all 16 perceptrons in the middle, and each keeps a weight for the 16 incoming connections. When we read the percentages for each color predicted by the network, we are just reading the output values from these 4 perceptrons.

Of course, if we have 48 color labels and 8 perceptrons per label, the middle layer is much wider, and we would have 48 output perceptrons. That's much harder to draw than it is simply to tell the neural network system the shape of what we want and let it do all the math. The essence of the network shape is still the same.

We'll leave for later a discussion of the actual weights and biases in a trained network. It would be nice to see the actual numbers involved and see what happens if, for instance, we present the network with 'burnt sienna' and see how it gets its prediction about that color.
